#4. How does MongoDB ensure high availability using replication?

Answer:

MongoDB uses replica sets to provide high availability.

A replica set contains:

Primary node: Receives all write operations.

Secondary nodes: Copy data from the primary through replication.

If the primary node fails:

The replica set automatically conducts an election.

One secondary becomes the new primary.

Applications continue working with minimal downtime.

This mechanism improves:

Fault tolerance

Data availability

Disaster recovery

#10. What are the ACID properties in the context of MongoDB transactions?

Answer:

MongoDB supports ACID transactions, ensuring reliable and consistent database operations.

The ACID properties are:

Atomicity: All operations in a transaction succeed together or fail together.
Consistency: The database remains in a valid state before and after the transaction.
Isolation: Transactions execute independently without interfering with one another.
Durability: Once a transaction is committed, the changes are permanently stored, even if the system crashes.

These properties ensure data integrity, especially in applications such as banking and e-commerce.

#16. What is the purpose of MongoDB's $lookup stage in aggregation?

Answer:

The $lookup stage is used to combine documents from two collections based on a common field.

It performs a join-like operation similar to SQL's JOIN.

Example:

```python
db.orders.aggregate([
   {
      $lookup: {
         from: "customers",
         localField: "customer_id",
         foreignField: "_id",
         as: "customer_details"
      }
   }
])
```

Benefits:

Combines data from multiple collections.
Reduces the need for multiple queries.
Simplifies reporting and analytics.
Helps retrieve related information in a single aggregation pipeline.

#9. Use aggregation to group orders by Region and calculate total sales per region.
Code:**bold text**

```python
from pymongo import MongoClient

# Connect to MongoDB
client = MongoClient("mongodb://localhost:27017/")
db = client["SuperstoreDB"]
collection = db["Orders"]

pipeline = [
    {
        "$group": {
            "_id": "$Region",
            "Total Sales": {"$sum": "$Sales"}
        }
    }
]

result = collection.aggregate(pipeline)

for record in result:
    print(record)
```

Explanation:
$group groups documents by the Region field.
$sum calculates the total sales for each region.
aggregate() executes the aggregation pipeline.

Expected Output (Sample):

```python
{'_id': 'Central', 'Total Sales': 501239.89}
{'_id': 'East', 'Total Sales': 678781.24}
{'_id': 'South', 'Total Sales': 391721.91}
{'_id': 'West', 'Total Sales': 725457.82}
```
